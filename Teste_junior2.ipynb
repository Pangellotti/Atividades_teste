```python
import requests
import pandas as pd
import json
```

```python
def tabela_pokemon(total_pokemons):


    '''
    Função para coletar atravez da pokeapi:
    Nome do Pokemon
    Habilidade 1
    Habilidade 2
    Habilidade 3
    Local aonde pode ser capturado

    Foi construida para coletar dados de quantos pokemons foram solicitados,
    como o objetivo era coletar os da primeira geração, o input recebe um valor
    e vai do pokemon 1 até esse valor, podendo ser até o 1.500.000

    o output é o dataframe montado
        
    '''

    df_nome= []
    df_habilidade1 = []
    df_habilidade2 = []
    df_habilidade3 = []
    df_local = []

    for i in range(1,total_pokemons+1):

        url = f"https://pokeapi.co/api/v2/pokemon/{i}"     #request do pokemon atravéz do Id
        pokemon = requests.get(url).json()
        nome_pokemon = pokemon["name"]                      #coleta o nome do pokemon
        habilidade = pokemon["abilities"]                   #chama o dicionario de habilidade do pokemon


        '''O looping funciona verifica se o pokemon possui a habilidade 1, 2 e 3, 
            caso não possua alguma, retorna "unk'''


        if len(habilidade) == 0:
            habilidade1 = "unk"
        else: 
            habilidade1 = pokemon["abilities"][0]["ability"]['name']

        if len(habilidade)==1:
            habilidade2 = "unk"
        else:
            habilidade2 = pokemon["abilities"][1]["ability"]['name']

        if len(habilidade)<=2:
            habilidade3 = "unk"
        else:
            habilidade3 = pokemon["abilities"][2]["ability"]['name']

        '''Verifica se tem o local disponivel, senão, retorna Unk'''
        
        url_local = pokemon["location_area_encounters"]
        local_get = requests.get(url_local).json()

        if len(local_get) == 0:
             local = "unk"
        else:
             local = local_get[0]['location_area']['name']

        df_nome.append(nome_pokemon)
        df_habilidade1.append(habilidade1)
        df_habilidade2.append(habilidade2)
        df_habilidade3.append(habilidade3)
        df_local.append(local)

    df= pd.DataFrame()
    df["Nome"] = df_nome
    df["habilidade_1"] = df_habilidade1
    df["habilidade_2"] = df_habilidade2
    df["habilidade_3"] = df_habilidade3
    df["local"] = df_local

    return df


        
       
```

```python
df = tabela_pokemon(150)
```
