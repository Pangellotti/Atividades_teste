#   Função

```python
def subsistemas(path_postos, path_vazão, subsistema):


    '''
    input: 
    path_postos: Local do arquivo de tabela dos postos
    path_vazão: local do arquivo de tabela de vazão
    subsistema: subsistema desejado ("N", "NE", "S" ou "SE")

    output: A primeira variavel recebe a média diaria, a segunda variavel a media mensal
   
    '''

    df_postos = pd.read_csv(path_postos)  #gera o df dos postos

    df_vazao = pd.read_csv(path_vazão) #gera o df de vazão

    df_total = pd.merge(df_postos, df_vazao, on=["num_posto"], how='left') #junta os 2 utilizando o codigo do posto como referencia

    df = df_total.loc[df_total.bin_ena == 1] #mantém apenas aonde bin_ena == 1

    df["ENA"] = df["val_produt"]*df["val_vaz_natr"] #gera a coluna ENA

    df.dat_medicao = pd.to_datetime(df.dat_medicao)

    if subsistema == "SE":

        df_SE = df.loc[df["nom_ssis"]=="SE"]   #separa apenas o subsitema do SE
        df_SE_daily_mean = df_SE.groupby("dat_medicao")["ENA"].mean()  #retonna a media diaria
        df_SE_monthly_mean=df_SE_daily_mean.resample("M").mean()      #retorna a media mensal

        return pd.DataFrame(df_SE_daily_mean), pd.DataFrame(df_SE_monthly_mean) #retonar no formato df

    elif subsistema == "S":

        df_S = df.loc[df["nom_ssis"]=="S"]
        df_S_daily_mean = df_S.groupby("dat_medicao")["ENA"].mean()
        df_S_monthly_mean = df_S_daily_mean.resample("M").mean()

        return  pd.DataFrame(df_S_daily_mean), pd.DataFrame(df_S_monthly_mean)

    elif subsistema == "NE":

        df_NE = df.loc[df["nom_ssis"]=="NE"]
        df_NE_daily_mean = df_NE.groupby("dat_medicao")["ENA"].mean()
        df_NE_monthly_mean = df_NE_daily_mean.resample("M").mean()

        return pd.DataFrame(df_NE_daily_mean), pd.DataFrame(df_NE_monthly_mean)
    
    elif subsistema == "N":

        df_N = df.loc[df["nom_ssis"]=="N"]
        df_N_daily_mean = df_N.groupby("dat_medicao")["ENA"].mean()
        df_N_monthly_mean = df_N_daily_mean.resample("M").mean()
        
        return pd.DataFrame(df_N_daily_mean), pd.DataFrame(df_N_monthly_mean)
        

```
